```python
# Plotting the Histogram
plt.hist(df['co2'], bins=40, color='yellow', edgecolor='black', label='bin width=10')

#Adding labels
plt.xlabel('CO2 Emissions')
plt.ylabel('Frequency')
plt.title('Distribution of CO2')

#Displaying the Histogram
plt.show()
```

Intrepretation of CO2 Histogram

This histogram depicts the distribution of CO2 emissions across vehicles. The x-axis represents the range of CO2 emissions, typically starting from 0 (least polluting) and increasing towards the right (more polluting). The y-axis represents the frequency, or the number of vehicles that emit a specific amount of CO2 within a certain range on the x-axis. From the histogram above, we see that the histogram of CO2 emissions shows a concentration of vehicle fuel efficiency in the range of 300 to 400.

```python
# Plotting the Histogram
plt.hist(df['feScore'], bins=10, color='skyblue', edgecolor='grey', label='bin width=50')

#Adding labels
plt.xlabel('feScore')
plt.ylabel('Frequency')
plt.title('Distribution of fuel-efficiency(fe)Score')

#Displaying the Histogram
plt.show()
```

Interpretation of the feScores Histogram


This histogram shows the distribution of the fuel-efficiency (fe) Score across the dataset. The x-axis represents the fe Score values ranging from 0 to 10, while the y-axis displays the frequency or count of observations (vehicles/models) for each fe Score value. The feScore histogram reveals a concentration of vehicles in the mid-range fuel efficiency category (4 to 6.5). A noteworthy peak at feScore 5 indicates a significant portion of the sample comprises vehicles with moderate fuel efficiency.  In simpler terms, a large proportion of vehicles exhibit a balanced fuel economy level

```python
# Plotting the Histogram
plt.hist(df['ghgScore'], bins=10, color='orange', edgecolor='purple', label='bin width=20')

#Adding labels
plt.xlabel('ghgScores')
plt.ylabel('Frequency')
plt.title('Distribution of ghgScore')

#Displaying the Histogram
plt.show()
```

```python
#Creating HeatMap, we need to print out the column with numeric values
#Printing out the first five column with numerical values only
df.select_dtypes('number').head()
```

```python
# Selecting specific columns with numerical values
selected_columns = ['id', 'year', 'cylinders', 'displ', 'pv2', 'pv4', 'city', 'UCity', 'highway', 'UHighway', 'comb', 'co2', 'feScore', 'ghgScore']

# Creating a subset DataFrame with selected columns
df1 = df[selected_columns]

# Calculating the correlation matrix
corr_df1 = df1.corr()

# Plotting the heatmap using Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(corr_df1, cmap='coolwarm', annot=True, fmt=".2f")
plt.title('Heatmap showing correlation of fuel-efficiency measures')
plt.show()
```

Interpretatipon of the HeatMap

Based on the heatmap, here is a thorough analysis of what is going on:

1) cylinders and displacement (0.93): This significant positive correlation indicates that as the number of cylinders in a vehicle increases, so does the engine size or displacement.

2) highway & city (0.92): This correlation shows that fuel-efficient vehicles on the highway are also fuel-efficient in city driving.

3) CO2 & cylinders (0.85), CO2 & displacement (0.86): There is a reasonably high correlation between CO2 emissions and engine cylinder displacement. This means that larger, more powerful engines emit more CO2.

4) feScore & city (0.91), feScore & highway (0.91), ghgScore & City (0.91), ghgScore & highway (0.9): These correlations indicate that the feScore and greenhouse gas Score (ghgScore) have significant positive correlations with city and highway fuel efficiency. Thus, higher scores in these categories indicate improved fuel efficiency.

5) city & combustion (0.99), feScore & combustion (0.93), and ghgScore & combustion (0.92): The highly high correlation between city and combustion, as well as the substantial correlations with feScore and ghgScore, indicate that these variables are closely related. This could be related to the fact that combustion is an important part of fuel economy, and thus city fuel efficiency, as measured by feScore and ghgScore.

6) ghgScore & feScore (0.99): This strong positive correlation indicates that vehicles with better fuel economy scores have fewer greenhouse gas emissions, supporting the notion that fuel-efficient vehicles are ecologically friendly.



In conclusion, these correlations indicate that some parameters, such as the number of cylinders, engine displacement, and combustion, are linked to both fuel efficiency and greenhouse gas emissions. These aspects are vital to consider while constructing more ecologically friendly vehicles
